```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
t, x1, y1, x2, y2, T, V = np.loadtxt("RungeKutta2_C/FoSM2_3b.dat", unpack=True)
#t2, y2 = np.loadtxt("FoSM2_2c.dat", unpack=True)

print(len(x1))
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.show()

E = T + V # energy E(t)
E0 = E[0]

plt.plot(t, ((E-E0)/E0))
plt.title("Relative Energy Error")
plt.ylabel("(E(t) - E0) / E0")
plt.xlabel("t")
plt.show()
```
